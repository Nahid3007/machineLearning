```python
import numpy as np
import matplotlib.pyplot as plt

from utils import (gradient_descent,
                   read_data)
```

# Example 

In `data/houses.txt` the target value `price ($1000)` of different houses depends on the features `size (sqft)`, `bedrooms`, `floors` and `age`.

```python
X_train, y_train = read_data('./data/houses.txt')

X_feature = ('size (sqft)','bedroom','floor','age')
```

Let's consider size as a single input feature and plot the data vs. the target value price.

## Plot size vs. price

```python
size  = X_train[:,0]
price = y_train

plt.figure()
ax = plt.axes()

ax.scatter(size,price,color='orange')

ax.set_xlabel(X_feature[0])
ax.set_ylabel("price (1000's)")
ax.set_title('Size vs. price')

plt.show()
```

## Linear Regression

In order to predict the price of a house of an arbitrary chosen size, a linear function can be tried to fit with the given data set using the gradient descent method.

```python
w,b,J_hist = gradient_descent(size,price,1000,1.0e-7)
```

## Optimized Parameter Values

```python
print(f'The results of the parameter values:')
print(f'w: {w}, b: {b}')
```

## Cost-Value vs. Iteration

```python
plt.figure()
ax = plt.axes()

ax.plot(np.arange(len(J_hist[:])),J_hist[:])

ax.set_xlabel('Number of Iterations')
ax.set_ylabel('Cost value')
ax.set_title("Cost vs. iteration")

plt.show()
```

## Results Linear Regression, Predicted Values

```python
y_pred = np.zeros(size.shape[0])
for i in range(size.shape[0]):
    y_pred[i] = np.dot(w,size[i]) + b
```

```python
plt.figure()
ax = plt.axes()

ax.scatter(size,price,color='orange', label = 'target')
ax.scatter(size,y_pred,color='blue', label = 'prediction')

ax.set_xlabel(X_feature[0])
ax.set_ylabel("price (1000's)")
ax.set_title('Size vs. price')

plt.legend(loc="upper left")

plt.show()
```

```python
for i in np.arange(0,len(price),20):
    print(f'Size: {size[i]:>4.0f}, Predicted value: {y_pred[i]:>4.1f}, target value: {price[i]:>4.1f}, rel. deviation [%]: {(100*((price[i]-y_pred[i])/y_pred[i])):>4.1f}')
```
